```python
import pandas as pd
import numpy as np

df = pd.read_csv('Superstore.csv',encoding = 'latin-1')
df.head()

print("shape",df.shape)
print("\n columns",df.columns.tolist())
print("\n Data types",df.dtypes)
print("\n Null values",df.isnull().sum())

print(df['Order Date'].head(10))

df['Order Date'] = pd.to_datetime(df['Order Date'], errors = 'coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors = 'coerce')


print("\n Dulpicate Values Before",df.duplicated().sum())
df = df.drop_duplicates()
print("\n Dulpicate Values After",df.duplicated().sum())

print(df['Order Date'].dtype)

df['Profit Margin %'] = round((df['Profit']/df['Sales'])*100,2)

df['Shipping Days']= (df['Ship Date'] - df['Order Date']).dt.days

df['loss'] = df['Profit']


```
