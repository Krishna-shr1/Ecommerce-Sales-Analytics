```python
import pandas as pd
import mysql.connector
from sqlalchemy import create_engine

conn = mysql.connector.connect(
    host = 'localhost',
    user = 'root',
    password = 'Krishna'
)

cursor = conn.cursor()
print("Connected to Mysql")

cursor.execute("CREATE DATABASE IF NOT EXISTS sales_project")
cursor.execute("USE sales_project")
print("Database Created")

engine = create_engine('mysql+mysqlconnector://root:Krishna@localhost/sales_project')
df= pd.read_csv('superstore_clean.csv')
df.to_sql('superstore', con = engine, if_exists= 'replace', index = False)

print("Data Loaded : Rows",len(df))

def run_query(query):
    return pd.read_sql(query,engine)

q1 = run_query("""SELECT Category, ROUND(SUM(profit),2) AS Total_profit, 
               ROUND(SUm(Sales),2) AS Total_Sales 
               FROM superstore
               GROUP BY Category 
               ORDER BY total_Profit 
               DESC""")

print("profit by category")
q1


q2 = run_query(""" SELECT `Product Name`, ROUND(SUM(profit),2) AS total_profit,
               ROUND(SUM(sales),2) AS total_sales 
               FROM superstore 
               GROUP BY `product name` 
               HAVING SUM(Profit) < 0
               ORDER  BY total_sales 
               DESC """)
print("High Sales but loss making producsts")
display(q2)


q3 =  run_query(""" SELECT State, ROUND(SUM(profit),2)AS total_profit, 
                ROUND(SUM(sales),2)AS total_sales 
                FROM superstore GROUP BY state 
                ORDER BY total_profit 
                DESC """)
print("top 10 Profitable states")
q3.head(10)


q4 = run_query(""" SELECT Segment, ROUND(SUM(sales),2) AS total_Sales,
               ROUND(SUM(Profit), 2) AS total_profit,
               COUNT(DISTINCT `Order ID`) AS total_orders
               FROM superstore 
               GROUP BY Segment 
               ORDER BY total_sales
               desc""")
print("Sales as per segment")
q4


q5 = run_query (""" SELECT `Ship Mode`, ROUND(AVG(`Shipping Days`),1) AS avg_shipping_days, 
                COUNT(*)AS total_orders 
                FROM superstore 
                GROUP BY `Ship Mode`
                ORDER BY avg_shipping_days 
                ASC""")
print("Average Shpping days by ship mode")
q5

q6 = run_query(""" SELECT `Order year`, `Order month`, ROUND(SUM(sales),2)AS total_sales, ROUND(SUM(profit),2)AS total_profit 
               FROM superstore
               GROUP BY `Order year`, `Order month`
               ORDER BY `Order year`, `Order month` """)
print("Sales and profit by month and year")
q6


q7 = run_query("""SELECT `Customer Name`, ROUND(SUM(sales),2)AS total_sales,ROUND(SUM(profit),2)AS total_profit FROM superstore
               GROUP BY `Customer Name`
               ORDER BY total_profit DESC
               limit 10 """)

print("Top 10 customers by profit")
q7

q8= run_query("""SELECT `Product Name`,ROUND(AVG(Discount),2)AS avg_discount, ROUND(SUM(Sales), 2) AS Total_Sales, ROUND(AVG(`Profit Margin %`),2)AS avg_margin,
              ROUND(SUM(Profit),2)AS Total_profit
              FROM superstore
              GROUP BY `Product Name`
              HAVING avg_discount >0.4
              ORDER BY Total_profit DESC
              limit 10 """)

print("Products with high discount and profit")
q8

q9 = run_query("""SELECT `Sub-Category`, ROUND(SUM(sales),2)AS total_sales, ROUND(SUM(profit),2)AS total_profit, ROUND(AVG(`Profit Margin %`),2)AS avg_margin
               FROM superstore
               GROUP BY `Sub-Category`
               ORDER BY avg_margin DESC """)
print("Profit margin by sub category")
q9
```
